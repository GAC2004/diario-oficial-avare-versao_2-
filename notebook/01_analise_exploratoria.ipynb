# Notebook 01 - Análise Exploratória dos Metadados

Este notebook realiza a leitura dos metadados, estatísticas básicas e distribuições por ano e tipo.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../data/metadata.csv')
display(df.head())
print(df.info())
print(df.describe(include='all'))

if 'ano_publicacao' in df.columns:
    df['ano_publicacao'].value_counts().sort_index().plot(kind='bar')
    plt.title('Publicações por Ano')
    plt.show()
```
