# Notebook 02 - Extração e Qualidade dos Textos

Analisa os textos extraídos e mede tamanho dos documentos.

```python
import json
import pandas as pd

with open('../data/textos_processados.json', encoding='utf-8') as f:
    docs=json.load(f)

df=pd.DataFrame(docs)
df['tamanho_texto']=df['texto'].str.len()
display(df[['tamanho_texto']].describe())
```
